```python
import matplotlib.pyplot as plt
```

```python
x = []
orm_insert = []
orm_extract = []
libpq_insert = []
libpq_extract = []

with open("plot.txt") as file:
    lines = file.readlines()
    for i in range(0, len(lines), 5):
        x.append(int(lines[i]))
        orm_insert.append(float(lines[i + 1]))
        orm_extract.append(float(lines[i + 2]))
        libpq_insert.append(float(lines[i + 3]))
        libpq_extract.append(float(lines[i + 4]))        
```

```python
plt.figure(figsize = (25, 10))
plt.plot(x, orm_insert, label = "Вставка ORM", color = "blue")
plt.plot(x, orm_extract, label = "Извлечение ORM", color = "red")
plt.plot(x, libpq_insert, label = "Вставка LIBPQ", color = "cyan")
plt.plot(x, libpq_extract, label = "Извлечение LIBPQ", color = "orange")
plt.xlabel("Количество элементов", fontsize="xx-large")
plt.xticks(fontsize="xx-large")
plt.ylabel("Время, с", fontsize="xx-large")
plt.yticks(fontsize="xx-large")
plt.legend(fontsize="xx-large")
plt.savefig("benchmark_plot.jpg")
plt.show()
```
